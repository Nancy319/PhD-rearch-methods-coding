```python
import pandas as pd
```

```python
citing_content = pd.read_csv('../data/Overton/covid_03_01/citing_content_overton_1.csv')
citing_content.info()
```

```python
# Open the file and read its contents into a list
with open('../data/Overton/covid_03_01/overton_policy_titles.txt', 'r') as file:
    policy_titles = file.readlines()

# Filter out any empty lines and whitespace
policy_filtered_titles = [title.strip() for title in policy_titles if title.strip()]

policy_filtered_titles[0:5]
```

```python
len(policy_filtered_titles)
```

```python
# Convert policy_filtered_titles to lowercase and strip whitespace
policy_filtered_titles = [title.lower().strip() for title in policy_filtered_titles]

# Convert citing_content['policy_doc_title'] to lowercase and strip whitespace
citing_content['policy_doc_title'] = citing_content['policy_doc_title'].str.lower().str.strip()

# Create a boolean mask indicating which rows of citing_content have a policy_doc_title that is in policy_filtered_titles
mask = citing_content['policy_doc_title'].isin(policy_filtered_titles)

# Use the mask to filter citing_content and save the result to matching_content
matching_content = citing_content[mask].reset_index()
matching_content.info()
```

```python
parliament_policy_extracted_info = pd.read_csv('../data/Overton/covid_03_01/11_unmatched_parliament_policy_info.csv')
```

```python
# Concatenate 'matching_content' and 'parliament_policy_extracted_info' vertically
science_policy_connection = pd.concat([matching_content, parliament_policy_extracted_info], ignore_index=True)
science_policy_connection.info()
```

```python
# Split the values in the policy_doc_area column of matching_content
split_areas = matching_content['policy_doc_area'].str.split(',')

# Remove any quotes around each area name
split_areas = [[area.strip("[]'") for area in areas] for areas in split_areas]

# Concatenate all of the split lists into a single list
all_areas = [area for areas in split_areas for area in areas]

# Use the set() function to get the unique values in the list
unique_areas = list(set(all_areas))
```

```python
unique_areas
```

```python
# Extract the last value after '>' for each area name (if applicable)
breakdown_policy_areas= []
for area in unique_areas:
    if '>' in area:
        breakdown_policy_area = area.split('>')[-1].strip().strip("'")
        breakdown_policy_areas.append(breakdown_policy_area)
    else:
        breakdown_policy_areas.append(area.strip().strip("'"))
```

```python
# Define a function to modify the policy_doc_area values as required
def modify_policy_area(area):
    # Split the area string by commas
    parts = area.split(',')
    
    # Strip any leading or trailing whitespace from each part
    parts = [part.strip() for part in parts]
    
    # Remove any symbols like '[]' or "'"
    parts = [part.replace('[', '').replace(']', '').replace("'", '') for part in parts]
    
    # Check each part to see if it matches the required structure
    for part in parts:
        if '>' in part:
            # Extract the last value after '>'
            modified_area = part.split('>')[-1]
            # Remove any symbols like '[]' or "'"
            modified_area = modified_area.replace('[', '').replace(']', '').replace("'", '')
            # Return the modified area value
            return modified_area
    
    # If none of the parts match the required structure, return the original area value
    return area

# Apply the modify_policy_area function to the policy_doc_area column and save the result to a new column
matching_content['modified_policy_area'] = matching_content['policy_doc_area'].apply(modify_policy_area)
```
