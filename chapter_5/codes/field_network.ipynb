```python
## Count the tokens of citing_paper_text
import nltk
from nltk.tokenize import word_tokenize

# Download the necessary resource for tokenization
nltk.download('punkt')

# Create a new list to store the tokenized text
tokenized_texts = []

# Tokenize the textual content
for text in science_policy_parliament['citing_paper_text']:
    if isinstance(text, str):
        tokens = word_tokenize(text)
    else:
        tokens = []
    tokenized_texts.append(tokens)

# Calculate total token count
total_token_count = sum(len(tokens) for tokens in tokenized_texts)

# Calculate average token count per cell
average_token_count = total_token_count / len(tokenized_texts)

print("Total token count:", total_token_count)
print("Average token count per cell:", average_token_count)
```

```python
# Create a new list to store the tokenized text
tokenized_parliament = []

# Tokenize the parliament discussion
for text in science_policy_parliament['parliament_related_discussion']:
    if isinstance(text, str):
        tokens = word_tokenize(text)
    else:
        tokens = []
    tokenized_parliament.append(tokens)

# Calculate total token count
total_parliament_token_count = sum(len(tokens) for tokens in tokenized_parliament)

# Calculate average token count per cell
average_parliament_token_count = total_parliament_token_count / len(tokenized_parliament)

print("Total parliament token count:", total_parliament_token_count)
print("Average paliament token count per cell:", average_parliament_token_count)
```

```python
with open('../data/chatgpt/iptc_topics_by_chatgpt35.json', 'r') as f:
    paper_iptc_topics = json.load(f)
```

```python
# Create a list of dictionaries from the JSON data
paper_iptc_topics_rows = []
for key, value in paper_iptc_topics.items():
    paper_iptc_topics_row = {
        'classified_text': key.split()[-1],
        'iptc_topics': value['iptc_topics']
    }
    paper_iptc_topics_rows.append(paper_iptc_topics_row)

# Convert the list of dictionaries into a pandas DataFrame
paper_iptc_topics_df = pd.DataFrame(paper_iptc_topics_rows)
```

```python
paper_iptc_topics_df.head()
```

```python
# Read the list from the file
with open('../data/chatgpt/citing_paper_text.txt', 'r') as f:
    paper_text_list = [line.strip() for line in f]

# Create a dictionary to map IDs to text
id_to_text = {str(i+1): paper_text_list[i] for i in range(len(paper_text_list))}

# Replace the IDs with the corresponding text in the DataFrame
paper_iptc_topics_df['classified_text'] = paper_iptc_topics_df['classified_text'].map(id_to_text)
```

```python
paper_iptc_topics_df.info()
```

```python
# Create a dictionary to map the classified text to the IPTC topics
classified_to_iptc = dict(zip(paper_iptc_topics_df['classified_text'], paper_iptc_topics_df['iptc_topics']))

# Add a new column to the science_policy_parliament DataFrame
science_policy_parliament['citing_paper_text_area'] = science_policy_parliament['citing_paper_text'].map(classified_to_iptc)
```

```python
# Split the values in policy_doc_area by >
field_connection['policy_doc_area'] = field_connection['policy_doc_area'].apply(lambda x: x.split('>'))

# Extract the last value in the resulting list
field_connection['policy_doc_area'] = field_connection['policy_doc_area'].apply(lambda x: x[-1])
```

```python
# Lowercase all values in the citing_paper_text_area column
field_connection['citing_paper_text_area'] = field_connection['citing_paper_text_area'].str.lower()
```

```python
# Drop the rows with None values
field_connection.dropna(inplace=True)

# Reset the index of the DataFrame
field_connection.reset_index(drop=True, inplace=True)
```

```python
gephi_data = field_connection[['citing_paper_text_area', 'policy_doc_area']]
gephi_data.columns = ['Source', 'Target']

# Save the gephi_data DataFrame as a CSV file to local storage
gephi_data.to_csv('../data/Gephi/field_connection_edge.csv', index=False)
```
